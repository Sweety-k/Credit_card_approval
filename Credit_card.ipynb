# **CREDIT CARD APPROVAL**

**Introduction**
A credit card is a payment card issued by a financial institution, usually a bank, that allows you to borrow money within a predefined credit limit. This borrowed amount can be used for purchasing goods and services or for withdrawing cash from ATMs.

**Key Features of Credit Cards:**

**Functionality:**

Borrowing Money: Credit cards let you borrow funds from the issuer up to your credit limit. This borrowed amount must be repaid, typically within a monthly billing cycle.

Interest Charges: If the balance is not paid in full by the due date, interest is charged on the remaining amount. The interest rate varies based on the card and your creditworthiness.

Repayment Options: Payments towards your credit card balance can be made through online banking, mobile apps, or by mail.

**Benefits:**

Convenience: Credit cards provide a convenient payment method, eliminating the need to carry cash. They are widely accepted by merchants both online and offline.

Rewards Programs: Many credit cards offer rewards programs that provide points, miles, or cash back for purchases. These rewards can be redeemed for travel, merchandise, or gift cards.

Building Credit: Responsible use of a credit card and timely bill payments help build a good credit history, facilitating easier access to loans, mortgages, and other forms of credit in the future.

**Drawbacks of Credit Cards:**

Debt Accumulation: Credit cards can lead to overspending and accumulating debt if not used carefully. This can result in high interest charges and potential financial difficulties.

Annual Fees: Certain credit cards, particularly those with extensive rewards programs, may charge annual fees.

High Interest Rates: Credit card interest rates can be quite high, making it crucial to pay off the balance in full each month to avoid incurring interest charges.

**Importance of This Project**

Credit card approval projects hold significant value for both customers and financial institutions. Here are key aspects to consider:

**For Financial Institutions:**

Reduced Risk: Accurate approval models help in lending to creditworthy customers, thereby minimizing defaults and financial losses.

Fairer Lending: Models can reduce bias, ensuring credit access based on objective criteria.

Improved Financial Inclusion: Expanding access to credit can promote financial health and well-being.

Personalized Offers: Models can tailor credit card options to individual needs and financial goals.

Increased Financial Security: Responsible credit card use can build a good credit history, providing access to better terms and rates in the future.

**Impact on the Banking Sector**

Banks receive numerous credit card applications, many of which are rejected due to factors such as high loan balances or low income levels. Manually analyzing these applications is time-consuming and prone to errors. Machine learning can automate this task, reducing human error and improving efficiency.

Handling Outliers
Outliers are data points in a dataset that deviate significantly from the majority of observations, being much higher or lower than the other values. In this context, since annual income is a crucial factor, we will address the outliers present in the annual income column by removing them.


The annual income distribution exhibits positive skewness with a skewness value of 2.15. To mitigate this skewness, we aim to explore various data transformation techniques such as square root, cube root, reciprocal, and logarithm to determine the most suitable method for improving the distribution's symmetry.

```python
sqr_root = np.sqrt(data['Annual_Income'])
cub_root = np.cbrt(data['Annual_Income'])
reciprocal = 1/data['Annual_Income']
log_tran = np.log(data['Annual_Income'])

plt.figure(figsize = (20,5))
plt.suptitle("Data Transformation of Annual Income")

plt.subplot(1,4,1)
sns.distplot(sqr_root, color = 'purple')
plt.title("Square root transformation")
print("Skew sqaure root:", sqr_root.skew().round(2))

plt.subplot(1,4,2)
sns.distplot(cub_root, color ='green')
plt.title("Cube Root Transformation")
print('Skew cube root;', cub_root.skew().round(2))

plt.subplot(1,4,3)
sns.distplot(reciprocal, color = 'brown')
plt.title("Reciprocal Transformation")
print("skew reciprocal:", reciprocal.skew().round(2))

plt.subplot(1,4,4)
sns.distplot(log_tran, color = 'red')
plt.title(" Log Transformation")
print("Skew Log tRnsformation:", log_tran.skew().round(2))
```

**The log transformation technique is the best as the skew= 0.06 which is almost a normal distribution curve**

Updating data['Annual_Income'] with log_tran is done to transform the original annual income data to a form that is closer to a normal distribution. This step is particularly useful when dealing with skewed data, as normality is often an underlying assumption in many statistical analyses and machine learning algorithms.

```python
data['Annual_Income'] = log_tran
data.head()
```

# **5. Machine Learning Model**

We will evaluate several machine learning algorithms to determine which one performs the best.

In machine learning, the task can be classified into two main types: classification or regression. In our case, we have a binary classification problem.

We will use the following algorithms:

1. Logistic Regression
2. Support Vector Machine (SVM)
3. Random Forest
4. Decision Tree
5. K-Nearest Neighbour (KNN)
6. Gradient Boost Classifier

For a binary classification problem, such as credit card approval, Type 1 errors (false positives) are particularly harmful. Therefore, we should prioritize precision over other metrics. High precision ensures that when the model predicts positive cases (credit card approvals), it is likely to be correct.

To focus on precision, we will use the parameters average='binary' and zero_division=1.

The machine learning algorithm with the highest precision score will be considered the best method.

```python
# we will import the accuracy, precision, confusion matrix and classification report for each ML method
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix, f1_score
```

# **1. Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# assign a logistic model to the variable **logistic_model**
logistic_model = LogisticRegression()

# train the model using the training datasets
logistic_model.fit(X_train_scaled, y_train)

# predict the test data
y_prediction_log = logistic_model.predict(X_test_scaled)

accuracy_log = accuracy_score(y_test, y_prediction_log)
precision_score_log = precision_score(y_test, y_prediction_log, average = 'binary', zero_division= 1)
recall_score_log = recall_score(y_test, y_prediction_log)
classification_report_log = classification_report(y_test, y_prediction_log)
confusion_matrix_log = confusion_matrix(y_test, y_prediction_log)
f1_score_log = f1_score(y_test, y_prediction_log)

print("Confusion Matrix :\n", confusion_matrix_log,"\n")
print("Accuracy : ", round(accuracy_log*100, 2),"%\n")
print("Precision Score :", round(precision_score_log*100, 2), "%\n")
print("Recall Score :", round(recall_score_log*100, 2),"%\n")
print("f1 score :", round(f1_score_log*100, 2),"%\n")
print(classification_report_log)

```

# **2. Random Forest Classifier**
